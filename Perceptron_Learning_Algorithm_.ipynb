```python
import numpy as np

data = pd.read_csv("/content/synthetic_dataset (2).csv")

# Rename the columns to 'x0', 'x1', and 'y'

# Add an extra column 'x0' with values of 1 for every row
data['x0'] = 1

# Extract features (x0, x1, and x2) and labels (y)
X = data[['x0', 'x1', 'x2']].values
y = data['y'].values

def perceptron_algorithm(X, y, max_iterations=100):
    w = np.zeros(X.shape[1])  # Initialize weight vector
    converged = False

    for _ in range(max_iterations):
        misclassified = 0
        for i in range(len(X)):
            if y[i] * np.dot(X[i], w) <= 0:
                w += y[i] * X[i]
                misclassified += 1
                break
        if misclassified == 0:
            converged = True
            break

    return w, converged


w_perceptron, _ = perceptron_algorithm(X, y)
print(_)
x1 = data['x1']
x2 = data['x2']
y = data['y']
colors = ['red' if label == 1 else 'blue' for label in y]
plt.scatter(x1, x2, c=colors)


plt.xlabel('x1-axis')
plt.ylabel('x2-axis')
plt.title('Data Points')
x1 = data['x1']
x2 = data['x2']
y = data['y']
colors = ['red' if label == 1 else 'blue' for label in y]
plt.scatter(x1, x2, c=colors)

# Plot the decision boundary as a line
x1_min, x1_max = x1.min() - 1, x1.max() + 1
x2_min, x2_max = x2.min() - 1, x2.max() + 1
xx, yy = np.meshgrid(np.arange(x1_min, x1_max, 0.01), np.arange(x2_min, x2_max, 0.01))
Z = np.dot(np.c_[np.ones_like(xx.ravel()), xx.ravel(), yy.ravel()], w_perceptron)
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, levels=[0], colors='k', linestyles='dashed')

plt.xlabel('x1-axis')
plt.ylabel('x2-axis')
plt.title('Data Points with Decision Boundary for PLA Algorithm')

plt.show()




```

```python
import numpy as np

data = pd.read_csv("/content/synthetic_dataset (2).csv")

# Rename the columns to 'x0', 'x1', and 'y'

# Add an extra column 'x0' with values of 1 for every row
data['x0'] = 1

# Extract features (x0, x1, and x2) and labels (y)
X = data[['x0', 'x1', 'x2']].values
y = data['y'].values

def pocket_algorithm(X, y, max_iterations=100):
    weights = np.zeros(X.shape[1])  # Initialize weight vector
    best_weights =np.zeros(X.shape[1])
    best_misclassifications = len(X)
    converged = False

    for _ in range(max_iterations):
        misClassifications = 0
        for i in range(len(X)):
            currentX = X[i]
            currentY = y[i]
            wTx = np.dot(currentX, weights)

            if currentY == 1 and wTx < 0:
                misClassifications += 1
                weights = weights + currentX
            elif currentY == -1 and wTx >= 0:
                misClassifications += 1
                weights = weights - currentX
        if misClassifications < best_misclassifications:
            best_misclassifications = misClassifications
            best_weights = weights.copy()
    return best_weights


w_perceptro = pocket_algorithm(X, y)
x1 = data['x1']
x2 = data['x2']
y = data['y']
colors = ['red' if label == 1 else 'blue' for label in y]
plt.scatter(x1, x2, c=colors)


plt.xlabel('x1-axis')
plt.ylabel('x2-axis')
plt.title('Data Points')
x1 = data['x1']
x2 = data['x2']
y = data['y']
colors = ['red' if label == 1 else 'blue' for label in y]
plt.scatter(x1, x2, c=colors)

# Plot the decision boundary as a line
x1_min, x1_max = x1.min() - 1, x1.max() + 1
x2_min, x2_max = x2.min() - 1, x2.max() + 1
xx, yy = np.meshgrid(np.arange(x1_min, x1_max, 0.01), np.arange(x2_min, x2_max, 0.01))
Z = np.dot(np.c_[np.ones_like(xx.ravel()), xx.ravel(), yy.ravel()], w_perceptro)
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, levels=[0], colors='k', linestyles='dashed')

plt.xlabel('x1-axis')
plt.ylabel('x2-axis')
plt.title('Data Points with Decision Boundary for Pocket Algorithm')

plt.show()




```
