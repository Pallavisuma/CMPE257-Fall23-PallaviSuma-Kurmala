```python
import pandas as pd
import numpy as np

data = pd.read_csv("/content/ZipDigits.train")

data
```

For Train Data Set

```python
import pandas as pd



df = pd.DataFrame(data)

def seperate_first_cloumn(row):
    split_values = row[0].split(maxsplit=1)
    return pd.Series(split_values)

split_df = df.apply(seperate_first_cloumn, axis=1)

# Rename the columns in the split DataFrame
split_df.columns = ['y', 'x']
split_df['y'] = split_df['y'].apply(lambda x: round(float(x)))

print(split_df)


```

```python
dig_for_bin_cls = [1, 5]

filtered_df = split_df[split_df['y'].isin(dig_for_bin_cls)]
filtered_df['y'] =filtered_df['y'].astype(str)

# Replace values in the 'y' column with -1 if they contain '5'
filtered_df['y'] = filtered_df['y'].apply(lambda x: '-1' if '5' in x else x)
filtered_df['y'] = filtered_df['y'].apply(lambda x: round(float(x)))

print(filtered_df)

def calculate_intensity(cell):
    values = [float(value) for value in cell.split()]
    return sum(values) / len(values)


filtered_df['intensity'] = filtered_df['x'].apply(calculate_intensity)


print(filtered_df[['x', 'intensity']])

def calculate_symmetry(cell):
    values = [float(value) for value in cell.split()]
    symmetry = sum(abs(values[i] - values[i + 1]) for i in range(len(values) - 1)) / len(values)
    return symmetry

filtered_df['symmetry'] = filtered_df['x'].apply(calculate_symmetry)

# Display the DataFrame with symmetry values
print(filtered_df[['x', 'intensity' ,'symmetry' ,'y']])
```

```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Separate data points based on class labels
class_1 = filtered_df[filtered_df['y'] == 1]
class_5 = filtered_df[filtered_df['y'] == -1]

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(class_1['intensity'], class_1['symmetry'], label='Class 1', color='blue')
plt.scatter(class_5['intensity'], class_5['symmetry'], label='Class 5', color='red')

# Label the axes and add a legend
plt.xlabel('Intensity')
plt.ylabel('Symmetry')
plt.legend(loc='best')

# Show the plot
plt.title('2D Scatter Plot of Intensity vs. Symmetry')
plt.show()

```
